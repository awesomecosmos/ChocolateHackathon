```python
for col in cocoa.columns:
    cocoa.loc[cocoa[col] == "Yes", col] = 1
    cocoa.loc[cocoa[col] == "No", col] = 0
    cocoa.loc[cocoa[col] == "Australia", col] = 0
    cocoa.loc[cocoa[col] == "Belgium", col] = 1
    cocoa.loc[cocoa[col] == "Canada", col] = 2
    cocoa.loc[cocoa[col] == "Ecuador", col] = 3
    cocoa.loc[cocoa[col] == "France", col] = 4
    cocoa.loc[cocoa[col] == "Italy", col] = 5
    cocoa.loc[cocoa[col] == "Other", col] = 6
    cocoa.loc[cocoa[col] == "U.S.A.", col] = 7
    cocoa.loc[cocoa[col] == "U.K.", col] = 8
    cocoa.loc[cocoa[col] == "Belize", col] = 9
    cocoa.loc[cocoa[col] == "Blend", col] = 10
    cocoa.loc[cocoa[col] == "Bolivia", col] = 11
    cocoa.loc[cocoa[col] == "Brazil", col] = 12
    cocoa.loc[cocoa[col] == "Colombia", col] = 13
    cocoa.loc[cocoa[col] == "Dominican Republic", col] = 14
    cocoa.loc[cocoa[col] == "Guatemala", col] = 15
    cocoa.loc[cocoa[col] == "Madagascar", col] = 16
    cocoa.loc[cocoa[col] == "Mexico", col] = 17
    cocoa.loc[cocoa[col] == "Nicaragua", col] = 18
    cocoa.loc[cocoa[col] == "Peru", col] = 19
    cocoa.loc[cocoa[col] == "Tanzania", col] = 20
    cocoa.loc[cocoa[col] == "Venezuela", col] = 21
    cocoa.loc[cocoa[col] == "Peru", col] = 22
    cocoa.loc[cocoa[col] == "Vietnam", col] = 23
```

```python
Choc = cocoa.drop(columns = "id")
```

```python
Choc[['company_location', 'country_of_bean_origin', 'sweet', 'nutty', 'cocoa', 'roasty', 'earthy', 'creamy', 'sandy', 'fatty', 'floral', 'intense', 'spicy', 'sour', 'vanilla', 'fruit', 'molasses', 'woody', 'sticky', 'coffee', 'rich', 'dried.fruit', 'ingredient_Beans', 'ingredient_Sugar', 'ingredient_Sweetener', 'ingredient_Cocoa_Butter', 'ingredient_vanilla ', 'ingredient_lecithin', 'ingredient_salt']] = Choc[['company_location', 'country_of_bean_origin', 'sweet', 'nutty', 'cocoa', 'roasty', 'earthy', 'creamy', 'sandy', 'fatty', 'floral', 'intense', 'spicy', 'sour', 'vanilla', 'fruit', 'molasses', 'woody', 'sticky', 'coffee', 'rich', 'dried.fruit', 'ingredient_Beans', 'ingredient_Sugar', 'ingredient_Sweetener', 'ingredient_Cocoa_Butter', 'ingredient_vanilla', 'ingredient_lecithin', 'ingredient_salt']].astype(int)
```

```python
Choc[['ingredient_vanilla']] = Choc[['ingredient_vanilla']].apply(pd.to_numeric, errors='coerce')
Choc[['ingredient_vanilla']] = Choc[['ingredient_vanilla']].fillna(0).astype(np.int64)
```
