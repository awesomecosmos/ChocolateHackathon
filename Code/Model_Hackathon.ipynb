```python
cocoa_train = pd.read_csv(r'/Users/manasvani/Documents/Chocolate_Hackathon/ChocolateCleanOfficial1.csv')
cocoa_test = pd.read_csv(r'/Users/manasvani/Documents/Chocolate_Hackathon/scoring_data_clean1.csv')
```

```python
cocoa_test = cocoa_test[['company_location', 'country_of_bean_origin', 'cocoa_percent', 'sweet',
       'nutty', 'cocoa', 'roasty', 'earthy', 'creamy', 'sandy', 'fatty',
       'floral', 'intense', 'spicy', 'sour', 'vanilla', 'fruit', 'molasses',
       'woody', 'sticky', 'coffee', 'rich', 'dried.fruit', 'ingredient_Beans',
       'ingredient_Sugar', 'ingredient_Sweetener', 'ingredient_Cocoa_Butter',
       'ingredient_lecithin', 'ingredient_salt',
       'number_of_ingredients', 'ingredient_vanilla']]
```

```python
amount_to_split = cocoa_test.shape[0]

X = cocoa_train.drop(['rating'], axis = 1)
Y = cocoa_train['rating']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=amount_to_split, random_state=4) 
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(X_train, Y_train);
# Use the forest's predict method on the test data
predictions = rf.predict(cocoa_test)
```

```python
# Calculate the absolute errors
errors = abs(predictions - Y_test)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / Y_test)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
print(predictions[:5])
```

```python
important = rf.feature_importances_
for i,v in enumerate(important):
    print('Feature: %0d, Score: %.5f' % (i,v))
```
